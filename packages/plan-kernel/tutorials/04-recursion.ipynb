# 04 — Recursion

A law's signature names `self` (slot 0) — the law itself. References to that name in the body produce self-application, and that's how recursion works in PLAN. No special `fix` operator: just refer to yourself.


## Factorial

`fact n = if n < 2 then 1 else n * fact(n - 1)`.

```plan
(#bind fact (#pin (#law "fact" (fact n)
  (If (Lt n 2) 1
      (Mul n (fact (Sub n 1)))))))
```

```plan
(fact 0)
```

```plan
(fact 1)
```

```plan
(fact 5)
```

```plan
(fact 10)
```

Reading the body:

* `(Lt n 2)` is the base condition.
* `(If c t e)` returns `t` if `c` is non-zero, else `e`.
* The recursive arm `(Mul n (fact (Sub n 1)))` calls `fact` directly — no special syntax for self-reference.

## Fibonacci

Naive recursion — each call branches into two:

```plan
(#bind fib (#pin (#law "fib" (fib n)
  (If (Lt n 2) n
      (Add (fib (Sub n 1)) (fib (Sub n 2)))))))
```

```plan
(fib 0)
```

```plan
(fib 1)
```

```plan
(fib 8)
```

```plan
(fib 12)
```

## A note on depth

The Marduk runtime evaluates by Python recursion. A self-recursive law of depth N takes O(N) Python stack frames (with a multiplier for evaluator bookkeeping). Toy examples like `(fib 12)` work fine; `(fib 30)` will exhaust the stack on default Python settings.

Production-grade depth handling is forward work — see the Marduk roadmap for native trampolining and jet support.

## A summing function

Tail-recursive shape: `sum n = sum_acc 0 n` where `sum_acc acc n` adds `n` into `acc` and recurses on `n-1`.

```plan
(#bind sum_acc (#pin (#law "sum_acc" (sum_acc acc n)
  (If (Eq n 0) acc
      (sum_acc (Add acc n) (Sub n 1))))))
```

```plan
(sum_acc 0 10)
```

```plan
(sum_acc 0 100)
```
