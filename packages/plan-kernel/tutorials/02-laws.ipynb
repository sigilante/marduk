# 02 — Laws

A **law** is PLAN's representation of a callable function. Three components: a name (any nat — usually a string-encoded one), an arity (how many arguments before saturation), and a body (an expression that computes the result).


## Surface syntax: `#law`

The macro `#law` constructs a law:

```
(#law "name" (self arg1 arg2 ...) body)
```

* The string after `#law` is the law's display name.
* The signature `(self arg1 ...)` declares the parameter names. `self` is the law itself (slot 0); `arg1` is slot 1, and so on.
* The body is an expression that can reference slots by name.

### Identity

Returning the argument unchanged:

````plan
(#law "id" (id x) x)
````

Apply it to a value:

````plan
((#law "id" (id x) x) 42)
````

## `#bind` — give a law a name

Inline laws are noisy. `#bind` ties a name in the top-level environment so subsequent cells can use it:

````plan
(#bind id (#law "id" (id x) x))
````

````plan
(id 99)
````

Bound names are visible across cells. `%env` shows them:

````plan
%env
````

## A doubler

Two-argument body: just call `Add` with the same input twice.

````plan
(#bind double (#law "double" (double x) (Add x x)))
````

````plan
(double 21)
````

## The K combinator

`K a b = a` — return the first argument, ignore the second. Useful as a constant function and in encoding booleans.

````plan
(#bind k (#law "k" (k a b) a))
````

````plan
(k 7 99)
````

## Composition

Functions compose by application. Define `quad` as `double` applied to `double`:

````plan
(#bind quad (#law "quad" (quad x) (double (double x))))
````

````plan
(quad 5)
````

## Pinning a law

`#pin` wraps a law in a Pin. The arity stays the same — applying a pinned law works identically to applying the law directly. Pins are how you mark a value as "finalized for sharing."

````plan
(#bind sq (#pin (#law "sq" (sq n) (Mul n n))))
````

````plan
(sq 9)
````
