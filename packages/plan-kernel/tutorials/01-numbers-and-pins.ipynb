# 01 — Numbers and pins

PLAN starts with one kind of value: the natural number. Type a nat, see it back.

This notebook teaches the smallest set of building blocks: bare nats, BPLAN arithmetic primitives, and pins. It assumes you have the `plan-kernel` Jupyter kernel installed and selected (``python -m plan_kernel install``).


## Bare nats

A bare nat is the number itself. The kernel echoes it back.

```plan
42
```

```plan
0
```

```plan
1024
```

## Arithmetic via BPLAN

PLAN proper has just three opcodes (Pin, Law, Elim). Arithmetic, comparison, and the rest of the usual primitives live in **BPLAN** — named operations dispatched via the pinned-nat dispatcher `<66>`. The surface syntax is ``(OpName arg1 arg2 ...)``.

```plan
(Add 2 3)
```

```plan
(Mul 7 6)
```

```plan
(Sub 10 3)
```

```plan
(Inc 41)
```

Compositions work the way you'd expect:

```plan
(Add (Mul 2 3) (Inc 4))
```

Saturating subtraction — `Sub` clamps at zero rather than going negative (PLAN nats are unsigned):

```plan
(Sub 3 10)
```

## Comparison

Booleans in PLAN are encoded as nats: `1` is true, `0` is false. `Eq`, `Lt`, `Gt`, etc. return one of those.

```plan
(Eq 5 5)
```

```plan
(Lt 3 7)
```

```plan
(Gt 3 7)
```

## Pins

A **pin** wraps a value. The notation `<v>` means "a pin "
around v". You construct one with `#pin`:

```plan
(#pin 5)
```

Pins are how PLAN does content addressing — two pins around structurally equal values are interchangeable. They're also how you mark a value as "finished, don't recompute". Most of the time you'll see them wrapping laws (the next lesson).
